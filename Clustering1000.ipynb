```python
import pandas as pd
from sklearn.cluster import KMeans

import parametrs
```

```python
data = pd.read_csv(parametrs.outpitFile,index_col=0)
```

```python
x = data.c0
y = data.c1
data = data.drop(['c0','c1'],axis=1)
```

```python
crystal_center = data.iloc[1]
gas_center = data.iloc[1000]
liquid_center = data.iloc[1700]
```

```python
len(data)
```

```python
data = data.fillna(0)
```

```python
data.head()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42).fit_predict(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,c=kmeans)
plt.ylabel('p')
#plt.yscale('log')
plt.xlabel('T')
#plt.ylim(0,10)
```

```python
plt.scatter(x,y,c=kmeans)
plt.ylim(.0,.01)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42).fit_predict(data)
```

```python
plt.scatter(x,y,c=kmeans)
plt.ylabel('p')
#plt.yscale('log')
plt.xlabel('T')
#plt.ylim(0,10)
```

```python
plt.scatter(x,y,c=kmeans)
plt.ylim(.0,.01)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
testModelAnswers = AgglomerativeClustering(n_clusters=5).fit_predict(data)
plt.scatter(x,y,c=testModelAnswers)
```

```python
plt.scatter(x,y,c=testModelAnswers)
plt.ylim(0,0.01)
```
